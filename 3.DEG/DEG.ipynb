```python
#import libraries
import numpy as np
import pandas as pd
```

```python
#Read the data
data = pd.read_csv('GSE130817_RNA_seq_raw_counts.csv.gz')
```

```python
data.tail()
```

```python
#Remove the geneId and save it to another dataset
gene = data['Unnamed: 0']
data.drop(columns = ['Unnamed: 0'],axis=1,inplace=True)
colo = data.columns
data.dropna(inplace = True)
```

```python
data
```

```python
#Get the cpm of the data
cpm = []
for i in range(0,len(data.columns)):
    tot = data.iloc[:,i].sum()
    count = (data.iloc[:,i]/tot)*10**6
    cpm.append(count)
```

```python
cpm = np.array(cpm)
cpm
#Convert the cpm to log values
log = np.log2(cpm+1)
```

```python
#Control sample in vec1 and Tumor sample in vec2
vec2 = log[0:4]
vec2
```

```python
vec1 = log[7:11]
vec1 = np.array(vec1)
vec1
len(vec1)
```

```python
meanTum = np.mean(vec1,axis=0)
meanCont = np.mean(vec2,axis=0)
```

```python
from scipy.stats import ttest_ind
Pvalue = ttest_ind(vec1,vec2).pvalue #Get the Pvalue
Pvalue
```

```python
log2FC = meanTum - meanCont #Get the log2FC values
log2FC
```

```python
res = np.stack((meanTum,meanCont,Pvalue,log2FC),axis = 1)
res
```

```python
#Convert to a Dataframe
data = pd.DataFrame(res) 
data = data
data
```

```python
#Add column names
data.columns = ['meanTumor','meanControl','Pvalue','Log2FC'] 
data
```

```python
#Fill na with 1
data['Pvalue'] = data['Pvalue'].fillna(1)
data
```

```python
#Drop all na vlaues
data.dropna(inplace = True)
```

```python
#Insert the gene column in the 1st Index
data.insert(0,'Gene',gene)
data
```

```python
#Plot the VolcanoPlot
import seaborn as sns
import matplotlib.pyplot as plt
from bioinfokit import visuz,analys
import matplotlib.pyplot as plt
visuz.GeneExpression.volcano(data, lfc='Log2FC', pv='Pvalue', 
    color=("#00239CFF", "grey", "#E10600FF"), plotlegend=True, legendpos='upper right', 
    legendanchor=(1.46,1))
```
