```python
#Calculate the cpm values from the data 
cpm = []
for i in range(0,len(data.columns)):
    tot = data.iloc[:,i].sum()
    count = (data.iloc[:,i]/tot)*10**6
    cpm.append(count)
    
```

```python
#Calculate the log values
log = np.log2(cpm+1)
```

```python
#Calculate zscore
import scipy.stats as stat
z=stat.zscore(log)
z
```

```python
#Convert to dataframe
df = pd.DataFrame(z).T
df = df.dropna(axis = 0)
df
```

```python
ar = df.values
#Calculate the Variance of the Zscore
var = []
for i in range(0,len(z)):
    vari = ar[:,i].var()
    var.append(vari)
    
```

```python
#Creating a function to calculate the log value from the csv file
import numpy as np

def lg(var):
    try:
        data = pd.read_csv(var)
        data.dropna(inplace = True)
        numerics = ['int16', 'int32', 'int64']
        data = data.select_dtypes(include=numerics)
        cpm = []
        for i in range(0,len(data.columns)):
            tot = data.iloc[:,i].sum()
            count = (data.iloc[:,i]/tot)*10**6
            cpm.append(count)
        cpm = np.array(cpm)
        log = np.log2(cpm+1)
        return log
    except:
        print("Enter the correct csv file")
    

    
```

```python
#Using the function
l = lg('GSE130817_RNA_seq_raw_counts.csv.gz')  
ld = pd.DataFrame(l).T
l
```

```python
#Convert the log values into a RDS files
import pyreadr as prd
prd.write_rds("Lg.Rds", ld)
```
